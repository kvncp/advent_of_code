```
import math

class Ship(object):
  def __init__(self):
    self.x = 0
    self.y = 0
    self.wx = 10
    self.wy = 1
    self.angle = 0

  def command(self, c, v):
    if c == 'N':
      self.wy += v
    elif c == 'S':
      self.wy -= v
    elif c == 'E':
      self.wx += v
    elif c == 'W':
      self.wx -= v
    elif c == 'L':
      self.wx, self.wy = (
          self.wx*math.cos(v*math.pi/180) - self.wy*math.sin(v*math.pi/180),
          self.wx*math.sin(v*math.pi/180) + self.wy*math.cos(v*math.pi/180))
    elif c == 'R':
      v = -v
      self.wx, self.wy = (
          self.wx*math.cos(v*math.pi/180) - self.wy*math.sin(v*math.pi/180),
          self.wx*math.sin(v*math.pi/180) + self.wy*math.cos(v*math.pi/180))
    elif c == 'F':
      self.x += v * self.wx
      self.y += v * self.wy 


s = Ship()


for r in directions:
  s.command(r[0], int(r[1:]))
  print(s.wx, s.wy, s.x, s.y)

print(abs(s.x) + abs(s.y))
```

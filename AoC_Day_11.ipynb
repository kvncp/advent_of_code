```
import copy
import itertools

class SeatMap(object):
  def __init__(self, seat_map):
    self.seat_map = [[x for x in row] for row in seat_map]
    self.new_seat_map = copy.deepcopy(self.seat_map)
    self.max_row = len(seat_map)
    self.max_col = len(seat_map[0])

  def get_adjacent_seats_occupied(self, row, col):
    occupied = 0
    for i in [-1, 0, 1]:
      for j in [-1, 0, 1]:
        if (i, j) == (0, 0):
          continue

        seat = self.find_visible_seat(row, col, i, j)
        if seat is None:
          continue

        if self.seat_map[seat[0]][seat[1]] == '#':
          occupied += 1
    return occupied

  def find_visible_seat(self, seat_row, seat_col, i, j):
    while True:
      seat_row = seat_row + i
      seat_col = seat_col + j

      if seat_row < 0 or seat_col < 0 or seat_row >= self.max_row or seat_col >= self.max_col:
        return None
      if self.seat_map[seat_row][seat_col] == '.':
        continue

      return seat_row, seat_col

  def eval_seat(self, row, col):
    if self.seat_map[row][col] == '.':
      return
    occupied = self.get_adjacent_seats_occupied(row, col)
    if self.seat_map[row][col] == 'L' and occupied == 0:
      self.new_seat_map[row][col] = '#'
    if self.seat_map[row][col] == '#' and occupied >= 5:
      self.new_seat_map[row][col] = 'L'

  def update_seats(self):
    for r in range(len(self.seat_map)):
      for c in range(len(self.seat_map[r])):
        self.eval_seat(r, c)

  def run_simulation(self):
    while True:
      self.update_seats()
      if self.seat_map == self.new_seat_map:
        break

      self.seat_map = copy.deepcopy(self.new_seat_map)

  def count_filled(self):
    return str(self).count('#')

  def __repr__(self):
    return '\n'.join(''.join(row) for row in self.seat_map)

sm = SeatMap(seats)
sm.run_simulation()
print(sm.count_filled())

print(sm)
```

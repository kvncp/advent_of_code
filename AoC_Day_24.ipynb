```
MOVES = {
    'e': (2, 0),
    'se': (1, -1),
    'sw': (-1, -1),
    'w': (-2, 0),
    'nw': (-1, 1),
    'ne': (1, 1),
}

def parse_input(moves):
  i = 0
  directions = []
  while i < len(moves):
    if moves[i] in ['e', 'w']:
      directions.append(moves[i])
      i += 1
    else:
      directions.append(moves[i:i+2])
      i += 2

  return directions

def walk(directions):
  x, y = (0, 0)
  for d in directions:
    xm, ym = MOVES[d]
    x, y = x + xm, y + ym

  return (x, y,)

flipped_tiles = set()
for row in puzzle:
  d = parse_input(row)
  flipped_tiles ^= set((walk(d),))

print(len(flipped_tiles))
```

```
def check_neighbors(black, target):
  total_black = 0
  for x, y in MOVES.values():
    if (target[0]+x, target[1]+y) in black:
      total_black += 1

  return total_black

def flip_white_tiles(black):
  new_black = set()
  seen = set()
  for tilex, tiley in black:
    for x, y in MOVES.values():
      candidate = tilex+x, tiley+y
      if candidate in black or candidate in seen:
        continue
      num_black = check_neighbors(black, candidate)
      if num_black == 2:
        new_black.add(candidate)
      seen.add(candidate)

  return new_black

def flip_black_tiles(black):
  new_black = set()
  for candidate in black:
    neighbors = check_neighbors(black, candidate)
    if neighbors == 1 or neighbors == 2:
      new_black.add(candidate)

  return new_black

for round in range(1, 101):
  flipped_tiles = flip_white_tiles(flipped_tiles) | flip_black_tiles(flipped_tiles)
  print(f'Day {round}: {len(flipped_tiles)}')
```

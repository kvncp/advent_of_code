```
# Part 1
class Math(object):
  def __init__(self, expression):
    self.expression = expression
    self.index = 0

  def evaluate(self):
    operation = '+'
    total = 0
    while self.index < len(self.expression):
      symbol = self.expression[self.index]
      self.index += 1
      if symbol == '(':
        total = self.operate(operation, total, self.evaluate())
      elif symbol[-1] == ')':
        return total
      elif symbol in ('*', '+'):
        operation = symbol
      else:
        total = self.operate(operation, total, symbol)
    
    return total

  @staticmethod
  def operate(symbol, x, y):
    if symbol == '+':
      return int(x) + int(y)
    elif symbol == '*':
      return int(x) * int(y)


sums = []
for row in examples.split('\n'):
  m = Math(row.replace(' ', ''))
  sums.append(m.evaluate())

print(sum(sums))
```

```
# Part 2
import functools


class Math2(object):
  def __init__(self, symbols):
    self.index = 0
    self.symbols = symbols
  
  def add_multiply(self):
    # print(' '.join(self.symbols[self.index:]))
    mul = []
    sums = []
    while self.index < len(self.symbols):
      symbol = self.symbols[self.index]
      self.index += 1
      if symbol == '*':
        mul.append(sum(sums))
        sums = []
      elif symbol == '+':
        continue
      elif symbol == '(':
        sums.append(self.add_multiply())
      elif symbol == ')':
        break
      else:
        sums.append(int(symbol))

    mul.append(sum(sums))
    return functools.reduce(lambda x, y: x*y, mul)


sums = []
for row in examples.split('\n'):
  m = Math2([x for x in row.replace(' ', '')])
  sums.append(m.add_multiply())

# print(sums)
print(sum(sums))
```

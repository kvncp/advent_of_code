```
# Problem #1.1
# O(N) Solution
values = set(input)
for x in input:
  if 2020 - x in values:
    print(x, 2020-x)
    print(x*(2020-x))
    break
```

```
# Problem #1.2
# O(N^2) Solution
values = set(input)
for x in input:
  for y in input:
    if 2020 - x - y in values:
      print(x, y)
      print(x * y * (2020-x-y))
      break
  else:
    continue
  break
```

```
ll = len(slope[0])


def hit_trees(slope, right, down):
  trees = 0
  pos = right
  for row in slope[down::down]:
    if row[pos] == '#':
      trees += 1
    pos = (pos + right) % ll

  return trees

print(hit_trees(slope, 3, 1))
```

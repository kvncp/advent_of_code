```python
def get_value(mm, k):
  for s, e in mm.keys():
    if k >= s and k < e:
      return mm[(s, e)] + (k-s)
  return k

def check_range(ranges, v):
  for s, e in ranges:
    if v >= s and v < s+e:
      return True
  return False
```
